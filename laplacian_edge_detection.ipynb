Laplacian Edge Detection

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread("C:/Users/ragha/Downloads/horse.webp")
res_img = cv.resize(img , (1000,600))
bil_img = cv.bilateralFilter(res_img , 8 , 60,60)

lap_img = cv.Laplacian(bil_img , cv.CV_64F , ksize = 3)

final  = np.uint8(np.absolute(lap_img))

cv.imshow("IMage" , final)
cv.waitKey(0)
cv.destroyAllWindows()


plt.subplot(2,1,1)
plt.imshow(final)
plt.axis("OFF")
plt.subplot(2,1,2)
img_ = cv.imread("C:/Users/ragha/Downloads/horse.webp")
img_ = cv.cvtColor(img_ , cv.COLOR_BGR2RGB)
plt.imshow(img_)
```

```python
video = cv.VideoCapture("C:/Users/ragha/Downloads/13074060_1920_1080_30fps.mp4")

while True:
    ret , frame = video.read()

    if not ret :
        break
    
    frame = cv.resize(frame , (1200,700))
    frame = cv.bilateralFilter(frame , 9 , 65 ,65)
    frame = cv.Laplacian(frame , cv.CV_64F , ksize= 3)
    frame = np.uint8(np.absolute(frame))
    cv.imshow("Video" , frame)
    
    plt.imshow(frame)
    if cv.waitKey(30) & 0xFF == ord('q'):
        break


video.release()
cv.destroyAllWindows()
```
