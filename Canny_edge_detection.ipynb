```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread("C:/Users/ragha/Downloads/tiger.jpeg")
image = cv.resize(image , (900,500))
image = cv.bilateralFilter(image , 8 , 65 , 65)
image = cv.Canny(image , 50 , 150)
cv.imshow("Cat" , image)
cv.waitKey(0)
cv.destroyAllWindows()

plt.subplot(2,1,1)
image =  cv.cvtColor(image , cv.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis("OFF")
plt.subplot(2,1,2)
image_ = cv.imread("C:/Users/ragha/Downloads/tiger.jpeg")
image_ =  cv.cvtColor(image_ , cv.COLOR_BGR2RGB)
plt.axis("OFF")
plt.imshow(image_)
```

```python
video = cv.VideoCapture("C:/Users/ragha/Downloads/13074060_1920_1080_30fps.mp4")

while True :
    ret , frame = video.read()

    if not ret :
        break


    frame = cv.resize(frame , (1200 , 700))
    frame = cv.bilateralFilter(frame , 8,65,65)
    frame = cv.Canny(frame , 50,150)

    cv.imshow("Playing" , frame)
    frame_ = cv.cvtColor(frame , cv.COLOR_BGR2RGB)
    plt.imshow(frame_)

    if cv.waitKey(30) & 0xFF == ord("q"):
        break

video.release()
cv.destroyAllWindows()


```
