```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt


image = cv.imread("C:/Users/ragha/Downloads/1.png")

sobel_x = cv.Sobel(image , cv.CV_64F , 1,0,ksize=3)
sobel_y = cv.Sobel(image , cv.CV_64F , 0,1, ksize = 3)

sobel_magnitude = np.sqrt(sobel_x**2 , sobel_y**2)
sobel_magnitude = np.uint8(sobel_magnitude)

_ , sobel_edges = cv.threshold(sobel_magnitude , 50,255 , cv.THRESH_BINARY)

cv.imshow("Edges" , sobel_edges)

cv.imshow("Numbers" , image)
cv.waitKey(0)
cv.destroyAllWindows()

plt.subplot(2,1,1)
plt.imshow(sobel_edges)
plt.axis("OFF")
plt.subplot(2,1,2)
plt.imshow(image)
```
